```python
# filter only to have a brand 'Innjoo'
df_innjoo = df[df['Brand'] == 'Innjoo']

# project dataframe with only 3 needed column
df_innjoo = df_innjoo[['RAM','Storage','Screen']]

# get the underlying Numpy array,
X = df_innjoo.values

# get the transpose of X
transpose_X = X.T

#compute matrix-matrix multiplication as XTX
XTX = transpose_X @ X

#compute the inverse of XTX
invert_XTX = np.linalg.inv(XTX)

#create and array y
y = np.array([1100, 1300, 800, 900, 1000, 1100])

# Calculate w by multiply the inverse of XTX with the transpose of X, and then multiply the result by y
w = (invert_XTX @ transpose_X) @ y

# calculate the sum of all the elements in w
w.sum()
```
